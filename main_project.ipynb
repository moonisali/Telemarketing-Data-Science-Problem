<h4>A)  Supervised Learning</h4><br>
<p  style="text-align:justify">

```python
job_house=pd.crosstab(df['job'],df['housing'],  aggfunc='count')
job_loan=pd.crosstab(df['job'],df['loan'],  aggfunc='count')
```

Why are we doing cross-tabulation of House and Loan with job.Our hypothesis is that values in the housing feature should be in the proportion of each job category. Therefore, using the cross-tabulation between house and job it will give us an idea of the proportion with which the unknown values will be filled.

```python
def job_house_fill(df,job_house):
    """Function for imputation via cross-tabulation to fill missing values for the 'housing' categorical feature"""
    jobs=df['job'].unique().tolist()
    house=["no","yes"]
    for j in jobs:
        ind=df[np.logical_and(np.array(df['housing']=='unknown'),np.array(df['job']==j))].index
        mask=np.random.rand(len(ind))<((job_house.loc[j]['no'])/(job_house.loc[j]['no']+job_house.loc[j]['yes']))
        ind1=ind[mask]
        ind2=ind[~mask]
        df.loc[ind1,"housing"]='no'
        df.loc[ind2,"housing"]='yes'
    return df
```

```python
def job_loan_fill(df,job_loan):
    """Function for imputation via cross-tabulation to fill missing values for the 'loan' categorical feature"""
    jobs=jobs=df['job'].unique().tolist()
    loan=["no","yes"]
    for j in jobs:
        ind=df[np.logical_and(np.array(df['loan']=='unknown'),np.array(df['job']==j))].index
        mask=np.random.rand(len(ind))<((job_loan.loc[j]['no'])/(job_loan.loc[j]['no']+job_loan.loc[j]['yes']))
        ind1=ind[mask]
        ind2=ind[~mask]
        df.loc[ind1,"loan"]='no'
        df.loc[ind2,"loan"]='yes'
    return df
```

```python
df=job_house_fill(df,job_house)
```

```python
df=job_loan_fill(df,job_loan)
```

We will replace the numerical variable 'pdays' with the categorical variable. The categories will be as follows:p_days_missing, pdays_less_5, pdays_bet_5_15, and pdays_greater_15.

```python
def pdays_handler(df):
    #Add new categorical variables to our dataframe.
    df['pdays_missing'] = 0
    df['pdays_less_5'] = 0
    df['pdays_greater_15'] = 0
    df['pdays_bet_5_15'] = 0
    df['pdays_missing'][df['pdays']==999] = 1
    df['pdays_less_5'][df['pdays']<5] = 1
    df['pdays_greater_15'][(df['pdays']>15) & (df['pdays']<999)] = 1
    df['pdays_bet_5_15'][(df['pdays']>=5)&(df['pdays']<=15)]= 1
    df_dropped_pdays = df.drop('pdays', axis=1, inplace=True)
    
```

Various packages used below are because of the following reasons:
- __time__ to evaluate the excution time.
- __cross_val_score__ for evaluation of cross validation accuracy
- __StratifiedKFold__ We will use this method for as cross validation parameter in the above function
- __train_test_split__ this will be used to split the data into test and train data.
- __accuracy_score__ this will be used to evaluate the general accuracy of the classifier

```python
import time
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

skf = StratifiedKFold(n_splits=20, random_state=0)

target = df['y']
features = df.drop('y',axis=1)

d_train, d_test, l_train, l_test = train_test_split(features, target,test_size=0.3, random_state=23)
```

#### 1) Desision Tree Classifier 
We will use this data in our Machine learning algorithm and print out the accuracy score as follows. Here we will use the Decision tree Classifier as shown below. We have used __StratifiedKFold()__ as cross validation method for checking the cross validation score.

```python
from sklearn.tree import DecisionTreeClassifier

start_time = time.time()

clf_ftr = DecisionTreeClassifier(criterion='entropy',min_samples_split = 9,max_depth=6,random_state=23)
clf_ftr = clf_ftr.fit(d_train, l_train)

predicted = clf_ftr.predict(d_test)
scores = cross_val_score(clf_ftr, features, target, cv=skf)

print("Decision Tree Classifier Accuracy: {0:.2%}".format(accuracy_score(predicted, l_test)))
print("Cross validation score: {0:.2%}".format(np.mean(scores)))
print("Execution time: {0:.5} seconds \n".format(time.time()-start_time))
```

## Feature Selection 
In this module we will use __Select by Model__ specifically Decision Tree Forest Method for selecting only those features which carry more information by using the __feature_importances__ method. This will help is reducing the time complexity as well as some undesirable noise. 

```python
#these two lists contain the feature score and feature name which contibutes to the specific score
ftr_scr =[]
ftr_info = []


for z in np.arange(0.001,.5,.001):
    imp = []
    #print(z)
    for i,x in zip(clf_ftr.feature_importances_,features.columns.values):
        if i >= z:            
            imp.append(i)
    ftr_scr.append(100*sum(imp))
    ftr_info.append(len(imp))

plt.figure(figsize=(12,8))
plt.scatter(x=ftr_scr,y=ftr_info, c="r", alpha=1, marker='o', s = 200)
plt.title("Feature vs Importance Graph")
plt.xlabel("Information gain")
plt.ylabel("Number of features")
plt.grid()
plt.show()
```

Now we will use that coefficient to select the minimum number of features with maximum information gain.Here we have used the coefficient of 0.001 and selected about 15 features which carry 99.48% information.

```python
imp = []
imp_cols = []
for i,x in zip(clf_ftr.feature_importances_,features.columns.values):
    
    if i >= 0.01:
        print(i,x)
        imp.append(i)
        imp_cols.append(x)
print(sum(imp))
print(len(imp))
```

```python
def convertInputFeatures(df):
    #Grouping Education variable
    df_basic_ed = df[(df.education == 'basic.4y') | (df.education == 'basic.6y') | (df.education == 'illiterate')]
    df_mid_ed = df[(df.education == 'basic.9y') | (df.education == 'high.school')]
    df_degree_ed = df[(df.education == 'professional.course') | (df.education == 'university.degree')]
     
    df.loc[df_basic_ed.index, 'education'] = 'Basic'
    df.loc[df_mid_ed.index, 'education'] = 'Mid'
    df.loc[df_degree_ed.index, 'education'] = 'Degree'
    
     
    df_unskilled = df[(df.job == 'blue-collar') | (df.job == 'housemaid')]
    df_service = df[(df.job == 'admin.') | (df.job == 'services') | (df.job == 'technician')]
    df_professional = df[(df.job == 'entrepreneur') | (df.job == 'self-employed') | (df.job == 'management')]
    df_student = df[(df.job == 'student')]
    df_retired = df[(df.job == 'retired')]
    df_unemployed = df[(df.job == 'unemployed')]      
        
    df.loc[df_unskilled.index, 'job'] = 'Unskilled'
    df.loc[df_service.index, 'job'] = 'Service'
    df.loc[df_professional.index, 'job'] = 'Professional'
    df.loc[df_student.index, 'job'] = 'Student'
    df.loc[df_retired.index, 'job'] = 'Retired'
    df.loc[df_unemployed.index, 'job'] = 'Unemployed'


    
    #Removing insignificant variables
    df.drop(['day_of_week'],axis=1,inplace=True)
    df.drop('duration',axis=1,inplace=True)
    
    df_loan_unknown = df[(df.loan == 'unknown')]
    df.loc[df_loan_unknown.index, 'loan'] = 'no'

    df_housing_unknown = df[(df.housing == 'unknown')]
    df.loc[df_housing_unknown.index, 'housing'] = 'yes'

    return df
```
